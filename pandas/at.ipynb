```python
import pandas as pd

# Leitura dos arquivos CSV
vendas_df = pd.read_csv('Vendas2017.csv', sep=';', encoding='ISO-8859-1')
produtos = pd.read_csv('CadastroProdutos.csv', sep=';', encoding='ISO-8859-1')
cliente = pd.read_csv('ClientesDF.csv', sep=';', encoding='ISO-8859-1')
lojas = pd.read_csv('Lojas.csv', sep=';', encoding='ISO-8859-1')

# Removendo colunas indesejadas do dataframe 'cliente'
cliente = cliente.drop(['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10'], axis=1, errors='ignore')

# Selecionando colunas específicas dos dataframes
cliente = cliente[['ÿID Cliente', 'E-mail']]
lojas = lojas[['ÿID Loja', 'Nome da Loja']]
produtos = produtos[['ID Produto', 'ï»¿Nome da Marca']]

# Renomeando colunas para corrigir problemas de codificação
cliente = cliente.rename(columns={'ÿID Cliente': 'ID Cliente'})
lojas = lojas.rename(columns={'ÿID Loja': 'ID Loja'})
produtos = produtos.rename(columns={'ï»¿Nome da Marca': 'Nome da Marca'})

# Juntando as tabelas com merge
vendas_df = vendas_df.merge(produtos, on='ID Produto')
vendas_df = vendas_df.merge(lojas, on='ID Loja')
vendas_df = vendas_df.merge(cliente, on='ID Cliente')
# Exibindo o dataframe final para verificação
frequencia_clientes = vendas_df['E-mail'].value_counts()
display(frequencia_clientes)

frequencia_clientes[:5].plot(figsize=(15,5), yticks=(range(0,100,10)))


```
